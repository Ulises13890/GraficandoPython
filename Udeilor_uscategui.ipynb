Genera un gráfico de dispersión las variables 𝑥 y 𝑦 = 𝑠𝑖𝑛(𝑥)+ ruido Gaussiano. Donde x es un array con númereos entre -2𝜋 𝑦 2𝜋. Grafica también los puntos 𝑦 = 𝑠𝑖𝑛(𝑥) en el mismo plot

```python
import numpy as np
import matplotlib.pyplot as plt

class GraficoDispersion:
    def __init__(self, num_puntos=100):
        self.num_puntos = num_puntos
        self.x = np.linspace(-2 * np.pi, 2 * np.pi, self.num_puntos)
        self.y = np.sin(self.x)
        self.y_con_ruido = self.y + np.random.normal(0, 0.1, self.num_puntos)

    def crear_grafico(self):
        plt.figure(figsize=(10, 6))
        
        # Graficar y = sin(x)
        plt.plot(self.x, self.y, label='y = sin(x)', color='blue', linewidth=2)
        
        # Graficar y = sin(x) + ruido Gaussiano
        plt.scatter(self.x, self.y_con_ruido, label='y = sin(x) + ruido Gaussiano', color='red', alpha=0.6)
        
        # Añadir etiquetas y título
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Gráfico de Dispersión: y = sin(x) con ruido Gaussiano')
        plt.legend()
        plt.grid(True)
        
        # Mostrar el gráfico
        plt.show()

# Crear una instancia de la clase y generar el gráfico
grafico = GraficoDispersion()
grafico.crear_grafico()
```

Utiliza la función np.meshgrid para crear una cuadrícula y luego aplica la función z = np.cos(x) + np.sin(y) para generar y mostrar un gráfico de contorno.

```python
# Crear una cuadrícula
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.linspace(-2 * np.pi, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)

# Aplicar la función z = cos(x) + sin(y)
Z = np.cos(X) + np.sin(Y)

# Crear el gráfico de contorno
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(contour)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de Contorno de z = cos(x) + sin(y)')
plt.show()
```

Crea un gráfico de dispersión con 1000 puntos aleatorios y utiliza la densidad de estos puntos para ajustar el color de cada punto.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generar 1000 puntos aleatorios
x = np.random.randn(1000)
y = np.random.randn(1000)

# Calcular la densidad de los puntos
xy = np.vstack([x, y])
z = gaussian_kde(xy)(xy)

# Crear el gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=z, s=50, edgecolor='none', cmap='viridis')
plt.colorbar(label='Densidad')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersión con Densidad de Puntos')
plt.show()
```

## Histogramas

Crea un histograma a partir de un array de 1000 números aleatorios generados con una distribución normal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar un array de 1000 números aleatorios con una distribución normal
data = np.random.normal(loc=0, scale=1, size=1000)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)

# Añadir etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Números Aleatorios con Distribución Normal')

# Mostrar el histograma
plt.show()
```

Genera dos sets de datos con distribuciones normales diferentes y muéstralos en el mismo histograma.

```python
# Generar dos sets de datos con distribuciones normales diferentes
data1 = np.random.normal(loc=0, scale=1, size=1000)  # Media 0, desviación estándar 1
data2 = np.random.normal(loc=2, scale=0.5, size=1000)  # Media 2, desviación estándar 0.5

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, alpha=0.5, label='Set de datos 1: µ=0, σ=1', color='blue')
plt.hist(data2, bins=30, alpha=0.5, label='Set de datos 2: µ=2, σ=0.5', color='red')

# Añadir etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histogramas de dos Sets de Datos con Distribuciones Normales Diferentes')
plt.legend()

# Mostrar el histograma
plt.show()
```
